```python
import pandas as pd
import matplotlib.pyplot as plt

# Loading the IPL dataset into a Pandas DataFrame
df = pd.read_csv("C:/Users/91904/Downloads/IPL_MATCHES.csv")
```

```python
# Counting the number of matches played in each venue
venue_counts = df['venue'].value_counts()

# Calculating the percentage of matches played in each venue
venue_percentages = 100 * venue_counts / venue_counts.sum()


# Ploting the pie chart
ax, fig= plt.subplots(figsize=(25,25))
plt.pie(venue_percentages, labels=venue_counts.index,autopct='%1.1f%%', textprops={'fontsize':18})
plt.title('Percentage of Matches Played in Each Venue')
plt.show()
```

```python
# Creating a new DataFrame to store the number of matches won and lost by each team
win_loss_df = pd.DataFrame(columns=['Team', 'Wins', 'Losses'])

# Iterating over the unique teams in the dataset
for team in df['team1'].unique():
    # Count the number of matches won by the team
    wins = len(df[df['winner'] == team])
    
    # Count the number of matches lost by the team
    losses = len(df[(df['team1'] == team) | (df['team2'] == team)]) - wins
    
    # Add a new row to the win_loss_df DataFrame
    win_loss_df = win_loss_df.append({'Team': team, 'Wins': wins, 'Losses': losses}, ignore_index=True)

    
    
# Ploting the bar graph
win_loss_df.plot.bar(x='Team', y=['Wins', 'Losses'], width=0.6)
plt.title('Number of Matches Won and Lost by Each Team')
plt.show()
```

```python
#plotting histogram to show the result margin distribution
ax, fig= plt.subplots(figsize=(8,10))
df['result_margin'].plot.hist(bins=10)
plt.title('Distribution of Result Margin')
plt.xlabel('Result Margin')
plt.show()
```
